## 🔢 Bagian 2: Pengenalan library "NumPy"

### Kenapa AI butuh NumPy?
AI sebenarnya cuma **operasi matematika dalam jumlah BESAR**. Bayangkan harus menjumlahkan 1 juta angka pakai kalkulator HP — bisa lembur seminggu! Dengan NumPy, hitungan jutaan angka selesai dalam hitungan detik. ⚡

### Konsep utama: **Array**
Array = deretan angka dalam satu wadah. Seperti deretan kursi di posyandu — diberi nomor, mudah dihitung.


```python
# Contoh: nilai ujian 5 anak kelas 6 SD
nilai = np.array([85, 90, 78, 92, 88])

print("Nilai siswi:", nilai)
print("Jumlah siswi:", len(nilai))
```

```python
# NumPy hitung otomatis — tidak perlu satu-satu!
print(f"Rata-rata    : {nilai.mean():.2f}")
print(f"Tertinggi    : {nilai.max()}")
print(f"Terendah     : {nilai.min()}")
print(f"Total        : {nilai.sum()}")
print(f"Selisih sebaran (std): {nilai.std():.2f}")
```

### 💡 Studi kasus mini: Tabungan PKK
Ada **6 ibu PKK** menabung selama **6 bulan** (dalam ribuan rupiah). Berapa total tabungan kelompok?


```python
# Setiap baris = 1 ibu; setiap kolom = 1 bulan
tabungan = np.array([
    [ 50,  75, 100,  60,  80,  90],   # Bu Ani
    [ 40,  60,  80,  50,  70,  85],   # Bu Budi
    [100, 120, 150, 110, 130, 140],   # Bu Citra
    [ 30,  45,  60,  40,  55,  70],   # Bu Dewi
    [ 80,  90, 100,  85,  95, 110],   # Bu Eka
    [ 60,  70,  80,  65,  75,  90],   # Bu Fitri
])

total_per_ibu   = tabungan.sum(axis=1)   # axis=1: jumlahkan per baris
total_per_bulan = tabungan.sum(axis=0)   # axis=0: jumlahkan per kolom

print("Total per ibu (ribu Rp)   :", total_per_ibu)
print("Total per bulan (ribu Rp) :", total_per_bulan)
print(f"\n💰 Total tabungan kelompok 6 bulan: Rp {tabungan.sum() * 1000:,}")
```

### 🔍 Filtering — memilih baris yang kita mau

Misalnya kita mau fokus ke **provinsi dengan IPM rendah** (di bawah 70).


```python
# Filter: provinsi dengan IPM 2022 di bawah 70
ipm_rendah = df[df['IPM_2022'] < 70]

print(f"📉 Ada {len(ipm_rendah)} provinsi dengan IPM 2022 di bawah 70:\n")
ipm_rendah[['Provinsi', 'IPM_2022', 'AKI_per_100rb', 'Melek_Huruf_Perempuan']]
```

```python
# Filter: provinsi dengan Angka Kematian Ibu di atas 200
aki_tinggi = df[df['AKI_per_100rb'] > 200].sort_values('AKI_per_100rb', ascending=False)

print("⚠️  Provinsi dengan Angka Kematian Ibu tertinggi:\n")
aki_tinggi[['Provinsi', 'AKI_per_100rb', 'IPM_2022']]
```

### 💭 Insight yang penting kita pahami

Lihat angka di atas: **AKI di Papua mencapai 565 per 100.000 kelahiran**, sementara di **DKI Jakarta hanya 55**. Artinya, **ibu hamil di Papua kira-kira 10× lebih berisiko** dibanding ibu di Jakarta.

Ini namanya **kesenjangan sosial**. Tugas kita yang sedang belajar sekarang — adalah membantu **menemukan, menjelaskan, dan mencari solusi** untuk kesenjangan seperti ini. 💪


---
### 🗂️ Grouping — meringkas data per kategori


```python
# Buat kolom baru: kategori IPM
def kategori_ipm(nilai):
    if nilai >= 80:   return 'Sangat Tinggi'
    elif nilai >= 70: return 'Tinggi'
    elif nilai >= 60: return 'Sedang'
    else:             return 'Rendah'

df['Kategori_IPM'] = df['IPM_2022'].apply(kategori_ipm)

# Ringkas per kategori
ringkasan = df.groupby('Kategori_IPM').agg({
    'Provinsi':              'count',
    'AKI_per_100rb':         'mean',
    'Melek_Huruf_Perempuan': 'mean'
}).round(1)

ringkasan.columns = ['Jumlah Provinsi', 'Rata-rata AKI', 'Rata-rata Melek Huruf (%)']
ringkasan
```

### 📌 Pola yang muncul

- Provinsi dengan IPM **Sangat Tinggi** → AKI **rendah** & melek huruf perempuan **tinggi**.
- Provinsi dengan IPM **Sedang** → AKI **tinggi** & melek huruf perempuan **lebih rendah**.

> **Kesimpulan praktis**: Pendidikan perempuan & kesehatan ibu **saling berkaitan erat**. Pola seperti ini nantinya bisa "dipelajari" otomatis oleh model AI. 🤖

### 🕳️ Menangani data kosong (*missing values*)

Di dunia nyata, **data sering tidak lengkap** — petugas lupa input, jaringan putus saat upload, dll. Pandas punya cara mudah menangani ini.


```python
# Contoh data dengan beberapa nilai kosong
data_posyandu = pd.DataFrame({
    'Nama_Posyandu':  ['Melati', 'Mawar', 'Anggrek', 'Tulip', 'Dahlia'],
    'Jumlah_Balita':  [45, 38, None, 52, 41],   # None = kosong
    'Petugas_Aktif':  [5, None, 4, 6, 5],
    'Bulan_Lapor':    ['Jan', 'Jan', 'Jan', None, 'Jan']
})

print("Data posyandu (dengan nilai kosong):")
data_posyandu
```

```python
# Deteksi: berapa banyak yang kosong di tiap kolom?
print("🔍 Jumlah nilai kosong per kolom:")
print(data_posyandu.isnull().sum())
```

```python
# Isi nilai kosong:
# - Untuk angka: pakai rata-rata atau median
# - Untuk teks : pakai 'Tidak Diketahui'

data_posyandu['Jumlah_Balita'] = data_posyandu['Jumlah_Balita'].fillna(
    data_posyandu['Jumlah_Balita'].mean()
)
data_posyandu['Petugas_Aktif'] = data_posyandu['Petugas_Aktif'].fillna(
    data_posyandu['Petugas_Aktif'].median()
)
data_posyandu['Bulan_Lapor'] = data_posyandu['Bulan_Lapor'].fillna('Tidak Diketahui')

print("✨ Setelah dibersihkan:")
data_posyandu
```

## 📈 Bagian 4 — Visualisasi: Mengubah Angka Jadi Gambar

> *"Satu gambar bisa bercerita lebih banyak dari ribuan angka."*

Sebelum AI bisa "belajar" dari data, **manusia** perlu melihat polanya dulu. Itu gunanya grafik.


```python
# Grafik 1 — Top 10 Provinsi dengan IPM tertinggi
top10_ipm = df.nlargest(10, 'IPM_2022').sort_values('IPM_2022')

plt.figure(figsize=(10, 6))
plt.barh(top10_ipm['Provinsi'], top10_ipm['IPM_2022'], color='#E91E63')
plt.xlabel('IPM 2022')
plt.title('🏆 10 Provinsi dengan IPM Tertinggi (2022)', fontsize=14, fontweight='bold')
plt.xlim(70, 85)

for i, v in enumerate(top10_ipm['IPM_2022']):
    plt.text(v + 0.1, i, f'{v}', va='center', fontsize=10)

plt.tight_layout()
plt.show()
```

```python
# Grafik 2 — Top 10 Provinsi dengan AKI tertinggi
top10_aki = df.nlargest(10, 'AKI_per_100rb').sort_values('AKI_per_100rb')
warna = ['#D32F2F' if x > 250 else '#F57C00' for x in top10_aki['AKI_per_100rb']]

plt.figure(figsize=(10, 6))
plt.barh(top10_aki['Provinsi'], top10_aki['AKI_per_100rb'], color=warna)
plt.xlabel('AKI per 100.000 Kelahiran')
plt.title('⚠️  10 Provinsi dengan Angka Kematian Ibu Tertinggi',
          fontsize=14, fontweight='bold')

for i, v in enumerate(top10_aki['AKI_per_100rb']):
    plt.text(v + 5, i, f'{int(v)}', va='center', fontsize=10)

plt.tight_layout()
plt.show()
```

```python
# Grafik 3 — Scatter: Hubungan IPM vs AKI
plt.figure(figsize=(10, 6))
sns.scatterplot(
    data=df, x='IPM_2022', y='AKI_per_100rb',
    hue='Kategori_IPM', s=150, palette='viridis', alpha=0.8
)

plt.title('🔬 Hubungan IPM dengan Angka Kematian Ibu',
          fontsize=14, fontweight='bold')
plt.xlabel('IPM 2022 (makin tinggi makin baik)')
plt.ylabel('Angka Kematian Ibu per 100.000')

# Tandai provinsi ekstrim
for _, row in df.iterrows():
    if row['AKI_per_100rb'] > 250 or row['IPM_2022'] > 80:
        plt.annotate(row['Provinsi'],
                     (row['IPM_2022'], row['AKI_per_100rb']),
                     fontsize=9, alpha=0.7,
                     xytext=(5, 5), textcoords='offset points')

plt.tight_layout()
plt.show()
```

### 🧠 Pola yang terlihat di scatter plot

- Titik-titik membentuk pola **kiri-atas ke kanan-bawah** → makin tinggi IPM, makin rendah AKI.
- Dalam statistik, ini disebut **korelasi negatif**.
- **Papua** terlihat sangat berbeda dari yang lain (di pojok kiri atas) → butuh intervensi serius.

> 💡 AI memprediksi pakai pola seperti ini. Misal: kalau kita tahu IPM suatu daerah, AI bisa **memperkirakan AKI**-nya dengan cukup akurat.


```python
# Grafik 4 — Tren IPM 2020–2022 untuk 5 provinsi terpilih
provinsi_pilihan = ['DKI Jakarta', 'DI Yogyakarta', 'Papua', 'NTT', 'Jawa Barat']
data_tren = df[df['Provinsi'].isin(provinsi_pilihan)]

plt.figure(figsize=(10, 6))
for _, row in data_tren.iterrows():
    plt.plot(['2020', '2021', '2022'],
             [row['IPM_2020'], row['IPM_2021'], row['IPM_2022']],
             marker='o', linewidth=2, markersize=8, label=row['Provinsi'])

plt.title('📈 Tren IPM 2020–2022', fontsize=14, fontweight='bold')
plt.xlabel('Tahun'); plt.ylabel('IPM')
plt.legend(loc='center left', bbox_to_anchor=(1.02, 0.5))
plt.grid(True, alpha=0.3)
plt.tight_layout()
plt.show()
```

```python
# Grafik 5 — Heatmap korelasi (peta panas)
kolom_numerik = ['IPM_2020', 'IPM_2021', 'IPM_2022',
                 'AKI_per_100rb', 'Melek_Huruf_Perempuan']
korelasi = df[kolom_numerik].corr()

plt.figure(figsize=(9, 7))
sns.heatmap(korelasi, annot=True, cmap='RdBu_r', center=0,
            fmt='.2f', square=True,
            cbar_kws={'label': 'Kekuatan Korelasi'})
plt.title('🔗 Korelasi Antar Variabel', fontsize=14, fontweight='bold')
plt.tight_layout()
plt.show()
```

### 📚 Cara baca heatmap

- **Angka mendekati 1** (biru tua) = naik bareng (korelasi positif kuat).
- **Angka mendekati −1** (merah tua) = berlawanan arah (korelasi negatif kuat).
- **Angka mendekati 0** (warna pucat) = tidak berhubungan.

**Yang kita pelajari dari heatmap di atas:**
- IPM ↔ AKI: korelasi **negatif kuat** (makin tinggi IPM, makin rendah AKI ✅).
- IPM ↔ Melek Huruf Perempuan: korelasi **positif** (mereka naik bareng).

## 🎯 Bagian 5 — Studi Kasus: "Cari 5 Provinsi Prioritas"

**Skenario**: Mbak/Ibu adalah analis di Kementerian PPPA. Pemerintah punya budget untuk **5 provinsi prioritas** yang paling butuh program kesehatan ibu & pendidikan perempuan.

**Tugas**: Pakai data untuk menentukan 5 provinsi tersebut, berdasarkan:
1. IPM **rendah**
2. AKI **tinggi**
3. Melek huruf perempuan **rendah**


```python
# Strategi: bikin skor gabungan untuk tiap provinsi
# Makin tinggi skor → makin butuh bantuan

# Normalisasi ke skala 0–100 supaya bisa dibandingkan
df['Skor_AKI']          = (df['AKI_per_100rb'] / df['AKI_per_100rb'].max()) * 100
df['Skor_IPM_rendah']   = (1 - df['IPM_2022'] / df['IPM_2022'].max()) * 100
df['Skor_Huruf_rendah'] = (1 - df['Melek_Huruf_Perempuan'] / 100) * 100

# Skor total = rata-rata 3 indikator
df['Skor_Prioritas'] = (df['Skor_AKI'] + df['Skor_IPM_rendah'] + df['Skor_Huruf_rendah']) / 3

prioritas = df.nlargest(5, 'Skor_Prioritas')[
    ['Provinsi', 'IPM_2022', 'AKI_per_100rb',
     'Melek_Huruf_Perempuan', 'Skor_Prioritas']
].round(1)

print("🎯 5 Provinsi Prioritas:\n")
prioritas
```

```python
# Visualisasikan hasilnya
prioritas_sorted = prioritas.sort_values('Skor_Prioritas')

plt.figure(figsize=(10, 6))
plt.barh(prioritas_sorted['Provinsi'],
         prioritas_sorted['Skor_Prioritas'],
         color='#C2185B')
plt.xlabel('Skor Prioritas (makin tinggi = makin butuh bantuan)')
plt.title('🎯 5 Provinsi Prioritas Intervensi', fontsize=14, fontweight='bold')

for i, v in enumerate(prioritas_sorted['Skor_Prioritas']):
    plt.text(v + 0.5, i, f'{v:.1f}', va='center', fontsize=11, fontweight='bold')

plt.tight_layout()
plt.show()

print("\n💡 Inilah cara data membantu pengambilan keputusan publik!")
```
